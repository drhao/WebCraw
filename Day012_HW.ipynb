## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

### ① 取出今天所有的發文

```python
#get URL
news_list = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    #print(d.find(class_="date").text, d.find_all('a')[-1].attrs['href'])
    article_url = "https://www.ettoday.net{}".format(d.find_all('a')[-1].attrs['href'])
    article_type = d.find('em').text
    
    t = requests.get(article_url)
    tSoup = BeautifulSoup(t.text, "html5lib")
    text = tSoup.find("div","story").text.strip()#.find_all('p')
    #print(article_type, text)
    news_list.append({
        'type': article_type,
        'text': text
    })
    
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
 news_list.append({
        'type': article_type,
        'text': text
    })
'''
import pandas as pd

art_text = []
art_type = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    #print(d.find(class_="date").text, d.find_all('a')[-1].attrs['href'])
    article_url = "https://www.ettoday.net{}".format(d.find_all('a')[-1].attrs['href'])
    article_type = d.find('em').text
    
    t = requests.get(article_url)
    tSoup = BeautifulSoup(t.text, "html5lib")
    text = tSoup.find("div","story").text.strip()#.find_all('p')
    #print(article_type, text)
    art_text.append(text)
    art_type.append(article_type)
```

### ③ 哪一個類別的文章最多

```python
article_all = pd.DataFrame(art_text, art_type).reset_index()
```
